#### 15596번

```python
def get_sum(a:list) -> int:
    s = 0
    for i in range(len(a)):
        s += a[i]
    return(s) 
```

#### 4673번

```python
def self_num(n):
    nums = range(1,n+1)
    not_self_list = []
    for i in range(1,n+1):
        if i<10:
            not_self_list.append(i+i)
        elif i<100:
            not_self_list.append(i+(i//10)+(i%10))
        elif i<1000:
            not_self_list.append(i+(i//100)+(i//10)+(i%10))
        else:
            not_self_list.append(i+(i//1000)+(i//100)+(i//10)+(i%10))
    return [x for x in nums if x not in not_self_list]
```

```python
self_num(10000)
```

#### 2675번

```python
t = int(input())
for j in range(t):
    n, s = input().split()
    n = int(n)
    for i in range(len(s)):
        print(s[i]*n, end='')
    print(end='\n')
```

```python
# 다른 풀이법 - 문자열 덧셈 성질 이용
t = int(input())
for j in range(t):
    n, s = input().split()
    text = ''
    for i in s:
        text += int(n)*i
    print(text)
```
